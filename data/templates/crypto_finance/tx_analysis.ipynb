```python
import random

# Basic Questions
def basic_gas_fee_calculation():
    gas_price = random.randint(10, 300)  # gwei
    gas_used = random.randint(21000, 100000)  # units
    amount_eth = random.uniform(0.1, 10)  # ETH
    conversion_rate = 1e9  # gwei to ETH
    total_fee_eth = (gas_price * gas_used) / conversion_rate
    question = (
        f"A transaction of {amount_eth:.2f} ETH uses {gas_used} gas units at a gas price of {gas_price} gwei. "
        "What is the total gas fee in ETH?"
    )
    solution = (
        f"Step 1: Calculate the total fee in gwei:\n"
        f"  {gas_price} gwei * {gas_used} units = {gas_price * gas_used} gwei.\n"
        f"Step 2: Convert gwei to ETH:\n"
        f"  Final Answer: {gas_price * gas_used} gwei / {conversion_rate} = {total_fee_eth:.8f} ETH."
    )
    return question, solution

def basic_balance_after_transaction():
    initial_balance = random.uniform(1, 10)  # BTC
    received = random.uniform(0.1, 5)  # BTC
    sent = random.uniform(0.1, 5)  # BTC
    final_balance = initial_balance + received - sent
    question = (
        f"A wallet starts with {initial_balance:.2f} BTC, receives {received:.2f} BTC, and sends {sent:.2f} BTC. "
        "What is the final balance of the wallet?"
    )
    solution = (
        f"Step 1: Add received amount:\n"
        f"  {initial_balance:.2f} + {received:.2f} = {initial_balance + received:.2f}.\n"
        f"Step 2: Subtract sent amount:\n"
        f"  Final Answer: {initial_balance + received:.2f} - {sent:.2f} = {final_balance:.2f}."
    )
    return question, solution

def basic_transaction_with_fee():
    initial_balance = random.uniform(1, 10)  # ETH
    sent = random.uniform(0.1, 5)  # ETH
    fee = random.uniform(0.001, 0.05)  # ETH
    final_balance = initial_balance - sent - fee
    question = (
        f"A wallet starts with {initial_balance:.2f} ETH and sends {sent:.2f} ETH, incurring a network fee of {fee:.4f} ETH. "
        "What is the final wallet balance?"
    )
    solution = (
        f"Step 1: Subtract sent amount:\n"
        f"  {initial_balance:.2f} - {sent:.2f} = {initial_balance - sent:.2f}.\n"
        f"Step 2: Subtract the network fee:\n"
        f"  Final Answer: {initial_balance - sent:.2f} - {fee:.4f} = {final_balance:.4f}."
    )
    return question, solution

def basic_transaction_fee_percentage():
    transaction_amount = random.uniform(0.1, 10)  # BTC
    fee_percentage = random.uniform(0.1, 5)  # %
    fee_btc = (transaction_amount * fee_percentage) / 100
    question = (
        f"A transaction of {transaction_amount:.2f} BTC incurs a fee of {fee_percentage:.2f}%. "
        "What is the fee in BTC?"
    )
    solution = (
        f"Step 1: Calculate the fee:\n"
        f"  Final Answer: {transaction_amount:.2f} * {fee_percentage:.2f} / 100 = {fee_btc:.8f} BTC."
    )
    return question, solution

# Intermediate Questions
def intermediate_conversion_rate():
    amount_usd = random.uniform(100, 5000)
    conversion_rate = random.uniform(30000, 60000)  # USD to BTC
    btc_received = amount_usd / conversion_rate
    question = (
        f"If 1 BTC is worth ${conversion_rate:.2f}, how much BTC would you receive for ${amount_usd:.2f}?"
    )
    solution = (
        f"Step 1: Divide the USD amount by the conversion rate:\n"
        f"  Final Answer: {amount_usd:.2f} / {conversion_rate:.2f} = {btc_received:.8f} BTC."
    )
    return question, solution

def intermediate_staking_rewards():
    staked_amount = random.uniform(1, 50)  # ETH
    annual_percentage = random.uniform(5, 20)  # %
    duration_years = random.randint(1, 5)
    total_rewards = staked_amount * (annual_percentage / 100) * duration_years
    question = (
        f"An investor stakes {staked_amount:.2f} ETH at an annual interest rate of {annual_percentage:.2f}% for {duration_years} years. "
        "What are the total staking rewards?"
    )
    solution = (
        f"Step 1: Calculate annual rewards:\n"
        f"  {staked_amount:.2f} * {annual_percentage:.2f} / 100 = {staked_amount * (annual_percentage / 100):.2f}.\n"
        f"Step 2: Multiply by the duration:\n"
        f"  Final Answer: {staked_amount * (annual_percentage / 100):.2f} * {duration_years} = {total_rewards:.2f} ETH."
    )
    return question, solution

def intermediate_multiple_transactions():
    initial_balance = random.uniform(1, 20)  # BTC
    tx1 = random.uniform(0.1, 5)  # BTC
    tx2 = random.uniform(0.1, 5)  # BTC
    fee = random.uniform(0.001, 0.05)  # BTC
    final_balance = initial_balance + tx1 - tx2 - fee
    question = (
        f"A wallet starts with {initial_balance:.2f} BTC, receives {tx1:.2f} BTC, and sends {tx2:.2f} BTC, with a network fee of {fee:.4f} BTC. "
        "What is the final wallet balance?"
    )
    solution = (
        f"Step 1: Add received amount:\n"
        f"  {initial_balance:.2f} + {tx1:.2f} = {initial_balance + tx1:.2f}.\n"
        f"Step 2: Subtract sent amount:\n"
        f"  {initial_balance + tx1:.2f} - {tx2:.2f} = {initial_balance + tx1 - tx2:.2f}.\n"
        f"Step 3: Subtract fee:\n"
        f"  Final Answer: {initial_balance + tx1 - tx2:.2f} - {fee:.4f} = {final_balance:.4f} BTC."
    )
    return question, solution

# Advanced Questions
def advanced_slippage_and_fee():
    trade_amount = random.uniform(100, 10000)  # USD
    slippage_percentage = random.uniform(0.1, 5)  # %
    fee_percentage = random.uniform(0.1, 3)  # %
    slippage_loss = (trade_amount * slippage_percentage) / 100
    fee = (trade_amount * fee_percentage) / 100
    total_cost = trade_amount + slippage_loss + fee
    question = (
        f"A trade worth ${trade_amount:.2f} incurs a slippage of {slippage_percentage:.2f}% and a fee of {fee_percentage:.2f}%. "
        "What is the total cost of the trade?"
    )
    solution = (
        f"Step 1: Calculate slippage loss:\n"
        f"  ${trade_amount:.2f} * {slippage_percentage:.2f} / 100 = ${slippage_loss:.2f}.\n"
        f"Step 2: Calculate fee:\n"
        f"  ${trade_amount:.2f} * {fee_percentage:.2f} / 100 = ${fee:.2f}.\n"
        f"Step 3: Add all costs:\n"
        f"  Final Answer: ${trade_amount:.2f} + ${slippage_loss:.2f} + ${fee:.2f} = ${total_cost:.2f}."
    )
    return question, solution

def advanced_compound_rewards():
    initial_stake = random.uniform(1, 100)  # ETH
    annual_percentage = random.uniform(5, 20)  # %
    years = random.randint(1, 5)
    final_amount = initial_stake * ((1 + (annual_percentage / 100)) ** years)
    question = (
        f"An investor stakes {initial_stake:.2f} ETH at an annual interest rate of {annual_percentage:.2f}% for {years} years, compounding annually. "
        "What is the final stake amount?"
    )
    solution = (
        f"Step 1: Use the compound interest formula:\n"
        f"  A = P * (1 + r)^t.\n"
        f"Step 2: Substitute values:\n"
        f"  A = {initial_stake:.2f} * (1 + {annual_percentage / 100:.2f})^{years}.\n"
        f"Step 3: Calculate final amount:\n"
        f"  Final Answer: A = {final_amount:.2f} ETH."
    )
    return question, solution

def advanced_multi_exchange_analysis():
    btc_initial = random.uniform(0.1, 5)  # BTC
    exchange_rate_1 = random.uniform(30000, 40000)  # USD/BTC
    exchange_rate_2 = random.uniform(25000, 35000)  # USD/BTC
    fee_percentage = random.uniform(0.1, 2)  # %
    usd_value = btc_initial * exchange_rate_1
    fee = (usd_value * fee_percentage) / 100
    btc_final = (usd_value - fee) / exchange_rate_2
    question = (
        f"A trader exchanges {btc_initial:.2f} BTC at an initial rate of ${exchange_rate_1:.2f}/BTC, then incurs a fee of {fee_percentage:.2f}%, and finally exchanges the remaining amount at ${exchange_rate_2:.2f}/BTC. "
        "How much BTC does the trader have after all transactions?"
    )
    solution = (
        f"Step 1: Calculate USD value after first exchange:\n"
        f"  {btc_initial:.2f} BTC * ${exchange_rate_1:.2f} = ${usd_value:.2f}.\n"
        f"Step 2: Calculate fee:\n"
        f"  ${usd_value:.2f} * {fee_percentage:.2f} / 100 = ${fee:.2f}.\n"
        f"Step 3: Subtract fee and convert back to BTC:\n"
        f"  Final Answer: (${usd_value:.2f} - ${fee:.2f}) / ${exchange_rate_2:.2f} = {btc_final:.8f} BTC."
    )
    return question, solution

# Example of generating all QA pairs
def generate_all_qa():
    methods = [
        basic_gas_fee_calculation,
        basic_balance_after_transaction,
        basic_transaction_with_fee,
        basic_transaction_fee_percentage,
        intermediate_conversion_rate,
        intermediate_staking_rewards,
        intermediate_multiple_transactions,
        advanced_slippage_and_fee,
        advanced_compound_rewards,
        advanced_multi_exchange_analysis
    ]

    for method in methods:
        question, solution = method()
        print(f"Question: {question}\nSolution: {solution}\n{'-' * 50}")

# Run the generator
generate_all_qa()
```
