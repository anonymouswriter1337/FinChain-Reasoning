```python
import random

company_names = [
    # Group 1: Innovation-inspired names
    "NextSphere", "VisionEdge", "InnoCore", "BrightVantage", "QuantumPeak",
    
    # Group 2: Nature-inspired names
    "EcoFlow", "BlueHorizon", "GreenCrest", "CrystalWave", "SolarisVentures",
    
    # Group 3: Tech-inspired names
    "CyberForge", "CodeCraft", "DigitalPulse", "NanoNetics", "OptiCore",
    
    # Group 4: Finance-inspired names
    "CapitalNest", "WealthHaven", "SecureFund", "PrimeAxis", "EquiBalance",
    
    # Group 5: Futuristic names
    "AstraNova", "NeoSphere", "OrionPath", "StarLine", "LumeTech",
    
    # Group 6: Abstract and modern names
    "Elevanta", "OmniLogic", "ZenithPoint", "NexaBridge", "Synervia",
    
    # Group 7: Professional-sounding names
    "ProVista", "CoreXpert", "AlphaStrive", "GlobalVantage", "VertexGrowth",
    
    # Group 8: Energy and motion-inspired names
    "MomentumWorks", "SparkDynamics", "FlowCatalyst", "EternaMotion", "VoltEdge",
    
    # Group 9: Minimalist names
    "AxisPoint", "KeyNode", "PrimeWave", "TrueHaven", "EchoBloom",
    
    # Group 10: Bold names
    "BoldWave", "PinnacleNet", "SummitForge", "VentureMark", "CrestTrail"
]


bank_names = ["JPMorgan Chase", "Bank of America", "Wells Fargo", "Goldman Sachs", "Morgan Stanley", "PNC Financial Services", "Capital One", "Barclays"]


# Template 1 (basic)
def template_calculate_net_cash_flow():
    """
    Scenario: A company reports cash inflows of $100,000 from operating activities, $50,000 from investing activities, 
    and $25,000 from financing activities. Additionally, the company has cash outflows of $40,000 for operating activities, 
    $10,000 for investing activities, and $5,000 for financing activities. Calculate the net cash flow for the company.
    """

    # Inputs
    company_name = random.choice(company_names)
    operating_inflows = random.randint(100000, 150000)
    investing_inflows = random.randint(50000, 100000)
    financing_inflows = random.randint(25000, 50000)
    operating_outflows = random.randint(40000, 60000)
    investing_outflows = random.randint(10000, 20000)
    financing_outflows = random.randint(5000, 10000)

    # Calculations
    net_operating = operating_inflows - operating_outflows
    net_investing = investing_inflows - investing_outflows
    net_financing = financing_inflows - financing_outflows
    net_cash_flow = net_operating + net_investing + net_financing

    question = (
        f"{company_name} reports cash inflows of ${operating_inflows} from operating activities, ${investing_inflows} from investing activities, "
        f"and ${financing_inflows} from financing activities. Additionally, the company has cash outflows of ${operating_outflows} for operating activities, "
        f"${investing_outflows} for investing activities, and ${financing_outflows} for financing activities. Calculate the net cash flow for {company_name}."
    )

    solution = f"""
    Step 1. Calculate net cash flow from operating activities:
            Net Operating Cash Flow = Operating Inflows - Operating Outflows
                                    = {operating_inflows} - {operating_outflows} = {net_operating}
    Step 2. Calculate net cash flow from investing activities:
            Net Investing Cash Flow = Investing Inflows - Investing Outflows
                                    = {investing_inflows} - {investing_outflows} = {net_investing}
    Step 3. Calculate net cash flow from financing activities:
            Net Financing Cash Flow = Financing Inflows - Financing Outflows
                                    = {financing_inflows} - {financing_outflows} = {net_financing}
    Step 4. Calculate total net cash flow:
            Total Net Cash Flow = Net Operating + Net Investing + Net Financing
                                = {net_operating} + {net_investing} + {net_financing} = {net_cash_flow}
    """

    return question, solution

# Template 2 (basic)
def template_calculate_closing_cash_balance():
    """
    Scenario: A company starts the month with a cash balance of $20,000. It has a net cash inflow of $30,000 from 
    operating activities, a net cash outflow of $10,000 from investing activities, and a net cash inflow of $5,000 
    from financing activities. Calculate the closing cash balance for the company.
    """

    # Inputs
    company_name = random.choice(company_names)
    opening_balance = random.randint(20000, 40000)
    net_operating_cash = random.randint(30000, 60000)
    net_investing_cash = random.randint(10000, 15000)
    net_financing_cash = random.randint(5000, 10000)

    # Calculations
    total_net_cash_flow = net_operating_cash - net_investing_cash + net_financing_cash
    closing_balance = opening_balance + total_net_cash_flow

    question = (
        f"{company_name} starts the month with a cash balance of ${opening_balance}. It has a net cash inflow of ${net_operating_cash} from "
        f"operating activities, a net cash outflow of ${net_investing_cash} from investing activities, and a net cash inflow of ${net_financing_cash} "
        f"from financing activities. Calculate the closing cash balance for the company."
    )

    solution = f"""
    Step 1. Calculate total net cash flow:
            Total Net Cash Flow = Net Operating Cash - Net Investing Cash + Net Financing Cash
                                = {net_operating_cash} - {net_investing_cash} + {net_financing_cash} = {total_net_cash_flow}
    Step 2. Calculate closing cash balance:
            Closing Cash Balance = Opening Balance + Total Net Cash Flow
                                 = {opening_balance} + {total_net_cash_flow} = {closing_balance}
    """

    return question, solution

# Template 3 (basic)
def template_calculate_changes_in_working_capital():
    """
    Scenario: A company reports an increase in accounts receivable by $10,000, an increase in accounts payable 
    by $5,000, and an increase in inventory by $15,000. Determine the change in working capital for the cash flow 
    statement.
    """

    # Inputs
    company_name = random.choice(company_names)
    accounts_receivable_change = random.randint(10000, 20000)
    accounts_payable_change = random.randint(5000, 10000)
    inventory_change = random.randint(15000, 30000)

    # Calculations
    change_in_working_capital = -accounts_receivable_change + accounts_payable_change - inventory_change

    question = (
        f"{company_name} reports an increase in accounts receivable by ${accounts_receivable_change}, an increase in accounts payable "
        f"by ${accounts_payable_change}, and an increase in inventory by ${inventory_change}. Determine the change in working capital for the cash flow "
        "statement."
    )

    solution = f"""
    Step 1. Impact of accounts receivable change:
            Change in Cash = -Increase in Accounts Receivable
                           = -{accounts_receivable_change}
    Step 2. Impact of accounts payable change:
            Change in Cash = +Increase in Accounts Payable
                           = +{accounts_payable_change}
    Step 3. Impact of inventory change:
            Change in Cash = -Increase in Inventory
                           = -{inventory_change}
    Step 4. Total change in working capital:
            Total Change = Step 1 + Step 2 + Step 3
                         = (-{accounts_receivable_change}) + {accounts_payable_change} + (-{inventory_change}) = {change_in_working_capital}
    """

    return question, solution

# Template 4 (basic)
def template_calculate_cash_paid_for_dividends():
    """
    Scenario: A company declares a total dividend of $50,000 during the year, of which $10,000 remains unpaid 
    at the year-end. Determine the cash paid for dividends as reported in the cash flow statement.
    """

    # Inputs
    company_name = random.choice(company_names)
    declared_dividends = random.randint(50000, 100000)
    unpaid_dividends = random.randint(10000, 20000)

    # Calculations
    cash_paid_for_dividends = declared_dividends - unpaid_dividends

    question = (
        f"{company_name} declares a total dividend of ${declared_dividends} during the year, of which ${unpaid_dividends} remains unpaid "
        "at the year-end. Determine the cash paid for dividends as reported in the cash flow statement."
    )

    solution = f"""
    Step 1. Calculate cash paid for dividends:
            Cash Paid for Dividends = Declared Dividends - Unpaid Dividends
                                    = {declared_dividends} - {unpaid_dividends} = {cash_paid_for_dividends}
    """

    return question, solution

# Template 5 (basic)
def template_calculate_tax_cash_outflow():
    """
    Scenario: A company reports an income tax expense of $25,000. At the beginning of the year, the company 
    had $5,000 in taxes payable, and at the end of the year, taxes payable increased to $8,000. Calculate the cash 
    outflow for taxes.
    """

    # Inputs
    company_name = random.choice(company_names)
    tax_expense = 25000
    opening_taxes_payable = random.randint(5000, 10000)
    closing_taxes_payable = random.randint(8000, 16000)

    # Calculations
    change_in_taxes_payable = closing_taxes_payable - opening_taxes_payable
    tax_cash_outflow = tax_expense - change_in_taxes_payable

    question = (
        f"{company_name} reports an income tax expense of ${tax_expense}. At the beginning of the year, the company "
        f"had ${opening_taxes_payable} in taxes payable, and at the end of the year, taxes payable increased to ${closing_taxes_payable}. Calculate the cash "
        "outflow for taxes."
    )

    solution = f"""
    Step 1. Calculate change in taxes payable:
            Change in Taxes Payable = Closing Taxes Payable - Opening Taxes Payable
                                    = {closing_taxes_payable} - {opening_taxes_payable} = {change_in_taxes_payable}
    Step 2. Calculate cash outflow for taxes:
            Tax Cash Outflow = Tax Expense - Change in Taxes Payable
                             = {tax_expense} - {change_in_taxes_payable} = {tax_cash_outflow}
    """

    return question, solution


# # Template 6 (intermediate - GPT4o)
# def template_calculate_working_capital_adjustment():
#     """
#     Scenario: A company needs to adjust its working capital for the year. It has an increase in accounts receivable of 
#     ${accounts_receivable}, an increase in inventory of ${inventory}, and an increase in accounts payable of 
#     ${accounts_payable}. Compute the net working capital adjustment for the year.
#     """

#     # Inputs
#     company_name = random.choice(company_names)
#     accounts_receivable = random.randint(20000, 100000)
#     inventory = random.randint(30000, 120000)
#     accounts_payable = random.randint(15000, 80000)

#     # Calculations
#     net_working_capital_adjustment = (accounts_receivable + inventory) - accounts_payable

#     question = (
#         f"{company_name} has an increase in accounts receivable of ${accounts_receivable}, an increase in inventory of ${inventory}, "
#         f"and an increase in accounts payable of ${accounts_payable}. Compute the net working capital adjustment."
#     )

#     solution = f"""
#     Step 1. Compute total working capital increases:
#             Total Increase = Accounts Receivable Increase + Inventory Increase
#             {accounts_receivable} + {inventory} = {accounts_receivable + inventory}
#     Step 2. Compute net working capital adjustment:
#             Net Adjustment = Total Increase - Accounts Payable Increase
#                            = ({accounts_receivable + inventory}) - {accounts_payable} = {net_working_capital_adjustment}
#     """

#     return question, solution


# Template 6 (intermediate - correction suggested by claude)
def template_calculate_working_capital_adjustment():
    """
    Scenario: A company needs to adjust its working capital for the year. It has an increase in accounts receivable of 
    ${accounts_receivable}, an increase in inventory of ${inventory}, and an increase in accounts payable of 
    ${accounts_payable}. Compute the net working capital adjustment for the year.
    """

    # Inputs
    company_name = random.choice(company_names)
    accounts_receivable = random.randint(20000, 100000)
    inventory = random.randint(30000, 120000)
    accounts_payable = random.randint(15000, 80000)

    # Calculations
    net_working_capital_adjustment = -(accounts_receivable + inventory) + accounts_payable

    question = (
        f"{company_name} has an increase in accounts receivable of ${accounts_receivable}, an increase in inventory of ${inventory}, "
        f"and an increase in accounts payable of ${accounts_payable}. Compute the net working capital adjustment."
    )

    solution = f"""
    Step 1. Compute total working capital increases:
            Total Increase = Accounts Receivable Increase + Inventory Increase
            {accounts_receivable} + {inventory} = {accounts_receivable + inventory}
    Step 2. Compute net working capital adjustment:
            Net Adjustment = -Total Increase + Accounts Payable Increase
                           = -({accounts_receivable + inventory}) + {accounts_payable} = {net_working_capital_adjustment}
    """

    return question, solution

# Template 7 (intermediate)
def template_calculate_cash_flow_from_financing_activities():
    """
    Scenario: A company has issued new equity worth ${equity_issuance} and raised ${debt_issuance} from long-term debt. 
    It has also repaid ${debt_repayment} in existing loans and paid ${dividends_paid} in dividends. Compute the net 
    cash flow from financing activities.
    """

    # Inputs
    company_name = random.choice(company_names)
    equity_issuance = random.randint(50000, 500000)
    debt_issuance = random.randint(100000, 700000)
    debt_repayment = random.randint(50000, 400000)
    dividends_paid = random.randint(20000, 150000)

    # Calculations
    net_cash_flow_financing = (equity_issuance + debt_issuance) - (debt_repayment + dividends_paid)

    question = (
        f"{company_name} issued new equity worth ${equity_issuance}, raised ${debt_issuance} from long-term debt, "
        f"repaid ${debt_repayment} in existing loans, and paid ${dividends_paid} in dividends. Compute the net "
        f"cash flow from financing activities."
    )

    solution = f"""
    Step 1. Compute total cash inflows from financing activities:
            Total Inflows = Equity Issuance + Debt Issuance
                          = {equity_issuance} + {debt_issuance} = {equity_issuance + debt_issuance}
    Step 2. Compute total cash outflows from financing activities:
            Total Outflows = Debt Repayment + Dividends Paid
                           = {debt_repayment} + {dividends_paid} = {debt_repayment + dividends_paid}
    Step 3. Compute net cash flow from financing activities:
            Net Cash Flow = Total Inflows - Total Outflows
                          = ({equity_issuance + debt_issuance}) - ({debt_repayment + dividends_paid}) = {net_cash_flow_financing}
    """

    return question, solution

# # Template 8 (intermediate - GPT4o)
# def calculate_free_cash_flow():
#     """
#     Scenario: A company reports an operating cash flow of ${operating_cash_flow}. It incurs capital expenditures of ${capex} 
#     and has changes in working capital amounting to ${working_capital_change}. Compute the company's free cash flow (FCF).
#     """

#     # Inputs
#     company_name = random.choice(company_names)
#     operating_cash_flow = random.randint(200000, 1000000)
#     capex = random.randint(50000, 300000)
#     working_capital_change = random.randint(-50000, 50000)

#     # Calculations
#     free_cash_flow = operating_cash_flow - capex + working_capital_change

#     question = (
#         f"{company_name} reports an operating cash flow of ${operating_cash_flow}. It incurs capital expenditures of ${capex} "
#         f"and has changes in working capital amounting to ${working_capital_change}. Compute the {company_name}'s free cash flow."
#     )

#     solution = f"""
#     Step 1. Compute free cash flow:
#             FCF = Operating Cash Flow - Capital Expenditures + Working Capital Changes
#                 = {operating_cash_flow} - {capex} + {working_capital_change} = {free_cash_flow}
#     """

#     return question, solution


# Template 8 (intermediate - correction suggested by claude)
def template_calculate_free_cash_flow():
    """
    Scenario: A company reports an operating cash flow of ${operating_cash_flow}. It incurs capital expenditures of ${capex} 
    and has changes in working capital amounting to ${working_capital_change}. Compute the company's free cash flow (FCF).
    """

    # Inputs
    company_name = random.choice(company_names)
    operating_cash_flow = random.randint(200000, 1000000)
    capex = random.randint(50000, 300000)
    working_capital_change = random.randint(-50000, 50000)

    # Calculations
    free_cash_flow = operating_cash_flow - capex

    question = (
        f"{company_name} reports an operating cash flow of ${operating_cash_flow}. It incurs capital expenditures of ${capex} "
        f"and has changes in working capital amounting to ${working_capital_change}. Compute the {company_name}'s free cash flow."
    )

    solution = f"""
    Step 1. Compute free cash flow:
            FCF = Operating Cash Flow - Capital Expenditures
                = {operating_cash_flow} - {capex} = {free_cash_flow}
    """

    return question, solution

# Template 9 (intermediate)
def template_calculate_cash_flow_impact_of_depreciation():
    """
    Scenario: A company records annual depreciation expenses of ${depreciation_expense} on its assets. The net income 
    reported is ${net_income}, and changes in working capital amount to ${working_capital_change}. Compute the operating 
    cash flow after adjusting for depreciation.
    """

    # Inputs
    company_name = random.choice(company_names)
    depreciation_expense = random.randint(20000, 150000)
    net_income = random.randint(50000, 500000)
    working_capital_change = random.randint(-50000, 50000)

    # Calculations
    operating_cash_flow = net_income + depreciation_expense + working_capital_change

    question = (
        f"{company_name} reports a net income of ${net_income} and annual depreciation expenses of ${depreciation_expense}. "
        f"It also has working capital changes of ${working_capital_change}. Compute the operating cash flow after adjusting "
        f"for depreciation."
    )

    solution = f"""
    Step 1. Add back non-cash depreciation expenses:
            Adjusted Net Income = Net Income + Depreciation Expense
                                = {net_income} + {depreciation_expense} = {net_income + depreciation_expense}
    Step 2. Adjust for working capital changes:
            Operating Cash Flow = Adjusted Net Income + Working Capital Change
                                = ({net_income + depreciation_expense}) + ({working_capital_change}) = {operating_cash_flow}
    """

    return question, solution

# Template 10 (intermediate)
def template_calculate_effective_tax_rate():
    """
    Scenario: A company reports an income before tax of ${income_before_tax} and has a total tax expense of ${tax_expense}. 
    Additionally, it receives tax credits amounting to ${tax_credits}. Compute the effective tax rate applied to the 
    company's earnings before tax.
    """

    # Inputs
    company_name = random.choice(company_names)
    income_before_tax = random.randint(100000, 1000000)
    tax_expense = random.randint(20000, 200000)
    tax_credits = random.randint(5000, 50000)

    # Calculations
    adjusted_tax_expense = tax_expense - tax_credits
    effective_tax_rate = (adjusted_tax_expense / income_before_tax) * 100

    question = (
        f"{company_name} reports an income before tax of ${income_before_tax} and has a total tax expense of ${tax_expense}. "
        f"Additionally, it receives tax credits amounting to ${tax_credits}. Compute the effective tax rate applied "
        f"to the {company_name}'s earnings before tax."
    )

    solution = f"""
    Step 1. Adjust the total tax expense by subtracting tax credits:
            Adjusted Tax Expense = Tax Expense - Tax Credits
                                 = {tax_expense} - {tax_credits} = {adjusted_tax_expense}
    Step 2. Compute the effective tax rate:
            Effective Tax Rate = (Adjusted Tax Expense / Income Before Tax) × 100
                               = ({adjusted_tax_expense} / {income_before_tax}) × 100 = {effective_tax_rate:.2f}%
    """

    return question, solution

# Template 11 (intermediate)
def template_calculate_operating_margin():
    """
    Scenario: A company generates total revenue of ${total_revenue} and incurs operating expenses of ${operating_expenses}. 
    Additionally, the company has depreciation and amortization expenses amounting to ${depreciation}. Compute the company's 
    operating margin percentage.
    """

    # Inputs
    company_name = random.choice(company_names)
    total_revenue = random.randint(500000, 5000000)
    operating_expenses = random.randint(200000, 4000000)
    depreciation = random.randint(50000, 300000)

    # Calculations
    operating_income = total_revenue - operating_expenses - depreciation
    operating_margin = (operating_income / total_revenue) * 100

    question = (
        f"{company_name} generates total revenue of ${total_revenue} and incurs operating expenses of ${operating_expenses}. "
        f"Additionally, {company_name} has depreciation and amortization expenses amounting to ${depreciation}. Compute the "
        f"{company_name}'s operating margin percentage."
    )

    solution = f"""
    Step 1. Compute the operating income:
            Operating Income = Total Revenue - Operating Expenses - Depreciation
                             = {total_revenue} - {operating_expenses} - {depreciation} = {operating_income}
    Step 2. Compute the operating margin:
            Operating Margin = (Operating Income / Total Revenue) × 100
                             = ({operating_income} / {total_revenue}) × 100 = {operating_margin:.2f}%
    """

    return question, solution

# Template 12 (advanced)
def template_calculate_levered_vs_unlevered_cash_flow():
    """
    Scenario: A company is analyzing its financial performance by comparing levered and unlevered free cash flow. 
    The net income of the company is ${net_income}. It incurs interest expenses of ${interest_expense}, 
    has depreciation and amortization of ${depreciation}, and changes in working capital amounting to ${working_capital_change}. 
    The company also reports capital expenditures of ${capex}. Compute the levered and unlevered free cash flows.
    """

    # Inputs
    company_name = random.choice(company_names)
    net_income = random.randint(50000, 200000)
    interest_expense = random.randint(5000, 30000)
    depreciation = random.randint(10000, 50000)
    working_capital_change = random.randint(-20000, 20000)
    capex = random.randint(20000, 100000)

    # Calculations
    fcff = net_income + interest_expense + depreciation - working_capital_change - capex
    fcfe = net_income + depreciation - working_capital_change - capex

    question = (
        f"{company_name} is analyzing its financial performance by comparing levered and unlevered free cash flow. "
        f"The net income of the company is ${net_income}, with interest expenses of ${interest_expense}, "
        f"depreciation and amortization of ${depreciation}, and changes in working capital amounting to {working_capital_change}. "
        f"{company_name} also reports capital expenditures of {capex}. Compute the levered and unlevered free cash flows."
    )

    solution = f"""
    Step 1. Compute unlevered free cash flow (FCFF):
            FCFF = Net Income + Interest Expense + Depreciation - Working Capital Change - CapEx
                 = {net_income} + {interest_expense} + {depreciation} - ({working_capital_change}) - {capex} = {fcff}
    Step 2. Compute levered free cash flow (FCFE):
            FCFE = Net Income + Depreciation - Working Capital Change - CapEx
                 = {net_income} + {depreciation} - ({working_capital_change}) - {capex} = {fcfe}
    """

    return question, solution

# Template 13 (advanced)
def template_calculate_cash_flow_impact_of_acquisition():
    """
    Scenario: A company is acquiring another firm and needs to evaluate the impact on its cash flow. 
    The acquisition cost is ${acquisition_cost}, and expected annual synergies from the acquisition are 
    ${synergies1}, ${synergies2}, ${synergies3}, and ${synergies4} for the next four years. The integration 
    costs in the first two years are ${integration_cost1} and ${integration_cost2}. Calculate the total net cash 
    flow impact of the acquisition over the four-year period.
    """

    # Inputs
    company_name = random.choice(company_names)
    acquisition_cost = random.randint(500000, 2000000)
    synergies1, synergies2, synergies3, synergies4 = [random.randint(50000, 200000) for _ in range(4)]
    integration_cost1, integration_cost2 = [random.randint(20000, 100000) for _ in range(2)]

    # Calculations
    total_synergies = synergies1 + synergies2 + synergies3 + synergies4
    total_integration_costs = integration_cost1 + integration_cost2
    net_cash_flow_impact = total_synergies - acquisition_cost - total_integration_costs

    question = (
        f"{company_name} is acquiring another firm for ${acquisition_cost}. It expects annual synergies of ${synergies1}, ${synergies2}, "
        f"${synergies3}, and ${synergies4} over the next four years. The integration costs in the first two years will be "
        f"${integration_cost1} and ${integration_cost2}. Compute the total net cash flow impact of the acquisition."
    )

    solution = f"""
    Step 1. Compute total synergy benefits:
            Total Synergies = Synergy1 + Synergy2 + Synergy3 + Synergy4
                            = {synergies1} + {synergies2} + {synergies3} + {synergies4} = {total_synergies}
    Step 2. Compute total integration costs:
            Total Integration Costs = Integration Cost1 + Integration Cost2
                                    = {integration_cost1} + {integration_cost2} = {total_integration_costs}
    Step 3. Compute net cash flow impact:
            Net Cash Flow Impact = Total Synergies - Acquisition Cost - Total Integration Costs
                                 = {total_synergies} - {acquisition_cost} - {total_integration_costs} = {net_cash_flow_impact}
    """

    return question, solution

# Template 14 (advanced)
def template_calculate_foreign_exchange_impact():
    """
    Scenario: A multinational corporation operates in multiple currencies and needs to evaluate the impact of 
    foreign exchange fluctuations on its cash flow. The company generates revenue of ${revenue_usd} in USD, 
    ${revenue_eur} in EUR, and ${revenue_gbp} in GBP. The exchange rates are 1 EUR = {eur_to_usd} USD and 
    1 GBP = {gbp_to_usd} USD. Compute the total cash flow impact in USD.
    """

    # Inputs
    company_name = random.choice(company_names)
    revenue_usd = random.randint(500000, 2000000)
    revenue_eur = random.randint(100000, 500000)
    revenue_gbp = random.randint(50000, 300000)
    eur_to_usd = random.uniform(1.05, 1.20)
    gbp_to_usd = random.uniform(1.25, 1.40)

    # Calculations
    total_revenue_usd = revenue_usd + (revenue_eur * eur_to_usd) + (revenue_gbp * gbp_to_usd)

    question = (
        f"A multinational corporation {company_name} generates revenue in multiple currencies: ${revenue_usd} in USD, "
        f"${revenue_eur} in EUR, and ${revenue_gbp} in GBP. The exchange rates are 1 EUR = {eur_to_usd} USD "
        f"and 1 GBP = {gbp_to_usd} USD. Compute the total revenue in USD after conversion."
    )

    solution = f"""
    Step 1. Convert EUR revenue to USD:
            EUR Revenue in USD = Revenue in EUR × Exchange Rate (EUR to USD)
                               = {revenue_eur} × {eur_to_usd} = {revenue_eur * eur_to_usd:.2f}
    Step 2. Convert GBP revenue to USD:
            GBP Revenue in USD = Revenue in GBP × Exchange Rate (GBP to USD)
                               = {revenue_gbp} × {gbp_to_usd} = {revenue_gbp * gbp_to_usd:.2f}
    Step 3. Compute total revenue in USD:
            Total Revenue in USD = Revenue in USD + Converted EUR Revenue + Converted GBP Revenue
                                 = {revenue_usd} + ({revenue_eur * eur_to_usd:.2f}) + ({revenue_gbp * gbp_to_usd:.2f}) = {total_revenue_usd:.2f}
    """

    return question, solution


# Template 15 (intermediate)
def template_calculate_cash_flow_sensitivity_analysis():
    """
    Scenario: A company wants to perform a sensitivity analysis on its net cash flow given fluctuating operating expenses. 
    The company has projected revenue of ${projected_revenue} and base operating expenses of ${base_operating_expenses}. 
    It considers three scenarios where expenses increase by {low_sensitivity}%, {medium_sensitivity}%, and {high_sensitivity}% respectively. 
    Compute the net cash flow under all three scenarios.
    """

    # Inputs
    projected_revenue = random.randint(500000, 2000000)
    base_operating_expenses = random.randint(200000, 800000)
    low_sensitivity = random.randint(5, 10)
    medium_sensitivity = random.randint(15, 20)
    high_sensitivity = random.randint(25, 30)

    # Calculations
    low_scenario_expenses = base_operating_expenses * (1 + low_sensitivity / 100)
    medium_scenario_expenses = base_operating_expenses * (1 + medium_sensitivity / 100)
    high_scenario_expenses = base_operating_expenses * (1 + high_sensitivity / 100)

    low_scenario_cash_flow = projected_revenue - low_scenario_expenses
    medium_scenario_cash_flow = projected_revenue - medium_scenario_expenses
    high_scenario_cash_flow = projected_revenue - high_scenario_expenses

    question = (
        f"A company projects revenue of ${projected_revenue} with base operating expenses of ${base_operating_expenses}. "
        f"It considers three scenarios where expenses increase by {low_sensitivity}%, {medium_sensitivity}%, and {high_sensitivity}%. "
        f"Compute the net cash flow under each scenario."
    )

    solution = f"""
    Step 1.   Compute operating expenses under each scenario:
    Step 1.1. Low Sensitivity: {base_operating_expenses} × (1 + {low_sensitivity}/100) = {low_scenario_expenses:.2f}
    Step 1.2  Medium Sensitivity: {base_operating_expenses} × (1 + {medium_sensitivity}/100) = {medium_scenario_expenses:.2f}
    Step 1.3  High Sensitivity: {base_operating_expenses} × (1 + {high_sensitivity}/100) = {high_scenario_expenses:.2f}
    Step 2.   Compute net cash flow for each scenario:
    Step 2.1  Low Sensitivity: {projected_revenue} - {low_scenario_expenses:.2f} = {low_scenario_cash_flow:.2f}
    Step 2.2  Medium Sensitivity: {projected_revenue} - {medium_scenario_expenses:.2f} = {medium_scenario_cash_flow:.2f}
    Step 2.3  High Sensitivity: {projected_revenue} - {high_scenario_expenses:.2f} = {high_scenario_cash_flow:.2f}
    """

    return question, solution

# Template 16 (advanced)
def template_calculate_cash_flow_forecasting():
    """
    Scenario: A company wants to forecast its cash flow for the next five years. It estimates annual revenue growth of 
    {growth_rate}% and expects its current revenue of ${initial_revenue} to increase accordingly. The company also assumes 
    operating expenses to be {expense_ratio}% of revenue and annual capital expenditures of ${capex}. Calculate the projected 
    net cash flow for each of the next five years.
    """

    # Inputs
    company_name = random.choice(company_names)
    initial_revenue = random.randint(500000, 2000000)
    growth_rate = random.randint(5, 15)
    expense_ratio = random.randint(50, 70)
    capex = random.randint(50000, 200000)
    num_years = random.randint(5, 10)

    # Calculations
    revenues = [initial_revenue * (1 + growth_rate / 100) ** i for i in range(1, num_years + 1)]
    expenses = [revenue * (expense_ratio / 100) for revenue in revenues]
    net_cash_flows = [revenue - expense - capex for revenue, expense in zip(revenues, expenses)]
    

    question = (
        f"{company_name} projects annual revenue growth of {growth_rate}%. It currently earns ${initial_revenue} in revenue. "
        f"Operating expenses are expected to be {expense_ratio}% of revenue, and annual capital expenditures are ${capex}. "
        f"Compute the projected net cash flow for the next {num_years} years."
    )

    solution = f"""
    Step 1. Compute projected revenue for each year:
            Revenue Year t = Initial Revenue × (1 + Growth Rate/100)^t
            {', '.join(f'Year {i+1}: {revenues[i]:.2f}' for i in range(num_years))}
    Step 2. Compute operating expenses:
            Operating Expenses = Revenue × Expense Ratio
            {', '.join(f'Year {i+1}: {revenues[i]:.2f} × {expense_ratio/100} = {expenses[i]:.2f}' for i in range(num_years))}
    Step 3. Compute net cash flow for each year:
            Net Cash Flow = Revenue - Operating Expenses - Capital Expenditures
            {', '.join(f'Year {i+1}: {revenues[i]:.2f} - {expenses[i]:.2f} - {capex} = {net_cash_flows[i]:.2f}' for i in range(num_years))}
    """

    return question, solution

if __name__ == "__main__":
    for idx, generate_fn in enumerate([
        template_calculate_net_cash_flow,
        template_calculate_closing_cash_balance,
        template_calculate_changes_in_working_capital,
        template_calculate_cash_paid_for_dividends,
        template_calculate_tax_cash_outflow,
        template_calculate_working_capital_adjustment,
        template_calculate_cash_flow_from_financing_activities,
        template_calculate_free_cash_flow,
        template_calculate_cash_flow_impact_of_depreciation,
        template_calculate_effective_tax_rate,
        template_calculate_operating_margin,
        template_calculate_levered_vs_unlevered_cash_flow,
        template_calculate_cash_flow_impact_of_acquisition,
        template_calculate_foreign_exchange_impact,
        template_calculate_cash_flow_sensitivity_analysis,
        template_calculate_cash_flow_forecasting
    ]):
        question, solution = generate_fn()
        print(f"Question {idx + 1}:", question)
        print("Solution:", solution)
        print("-" * 50) 
```
