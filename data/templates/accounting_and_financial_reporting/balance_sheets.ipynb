```python
import random


company_names = [
    # Group 1: Innovation-inspired names
    "NextSphere", "VisionEdge", "InnoCore", "BrightVantage", "QuantumPeak",
    
    # Group 2: Nature-inspired names
    "EcoFlow", "BlueHorizon", "GreenCrest", "CrystalWave", "SolarisVentures",
    
    # Group 3: Tech-inspired names
    "CyberForge", "CodeCraft", "DigitalPulse", "NanoNetics", "OptiCore",
    
    # Group 4: Finance-inspired names
    "CapitalNest", "WealthHaven", "SecureFund", "PrimeAxis", "EquiBalance",
    
    # Group 5: Futuristic names
    "AstraNova", "NeoSphere", "OrionPath", "StarLine", "LumeTech",
    
    # Group 6: Abstract and modern names
    "Elevanta", "OmniLogic", "ZenithPoint", "NexaBridge", "Synervia",
    
    # Group 7: Professional-sounding names
    "ProVista", "CoreXpert", "AlphaStrive", "GlobalVantage", "VertexGrowth",
    
    # Group 8: Energy and motion-inspired names
    "MomentumWorks", "SparkDynamics", "FlowCatalyst", "EternaMotion", "VoltEdge",
    
    # Group 9: Minimalist names
    "AxisPoint", "KeyNode", "PrimeWave", "TrueHaven", "EchoBloom",
    
    # Group 10: Bold names
    "BoldWave", "PinnacleNet", "SummitForge", "VentureMark", "CrestTrail"
]

bank_names = ["JPMorgan Chase", "Bank of America", "Wells Fargo", "Goldman Sachs", 
              "Morgan Stanley", "PNC Financial Services", "Capital One", "Barclays"]

# Template 1 (basic)
def template_cash_accounts_payable():
    """
    Scenario: Cash and Accounts Payable
    """
    company_name = random.choice(company_names)
    cash = random.randint(5000, 20000)
    accounts_payable = random.randint(2000, 10000)
    net_cash = cash - accounts_payable

    question = (
        f"{company_name} has ${cash} in cash and ${accounts_payable} in accounts payable. Calculate the net cash position "
        f"of the company after paying off all its accounts payable."
    )

    solution = (
        f"Step 1. Net Cash Position = Cash - Accounts Payable\n"
        f"                          = {cash} - {accounts_payable} = ${net_cash}"
    )

    return question, solution

# Template 2 (basic)
def template_balance_sheet_equation():
    """
    Scenario: Simple Balance Sheet Equation
    """
    company_name = random.choice(company_names)
    assets = random.randint(10000, 50000)
    liabilities = random.randint(5000, 30000)
    equity = assets - liabilities

    question = (
        f"{company_name} has ${assets} in assets and ${liabilities} in liabilities. Using the balance sheet equation, "
        f"calculate the company’s equity."
    )

    solution = (
        f"Step 1. Equity = Assets - Liabilities\n"
        f"               = {assets} - {liabilities} = ${equity}"
    )

    return question, solution

# Template 3 (basic)
def template_depreciation_fixed_asset():
    """
    Scenario: Depreciation of a Fixed Asset
    """
    company_name = random.choice(company_names)
    purchase_price = random.randint(20000, 100000)
    annual_depreciation = random.randint(2000, 10000)
    years = random.randint(2, 10)
    book_value = purchase_price - (annual_depreciation * years)

    question = (
        f"{company_name} purchased a piece of machinery for ${purchase_price}. The machinery is expected to depreciate "
        f"by ${annual_depreciation} annually. Calculate the book value of the machinery after {years} years."
    )

    solution = (
        f"Step 1. Book Value = Purchase Price - (Annual Depreciation * Years)\n"
        f"                   = {purchase_price} - ({annual_depreciation} * {years}) = ${book_value}"
    )

    return question, solution

# Template 4 (basic)
def template_inventory_closing():
    """
    Scenario: Inventory Valuation
    """
    company_name = random.choice(company_names)
    opening_inventory = random.randint(10000, 50000)
    purchases = random.randint(5000, 30000)
    sales = random.randint(5000, 30000)
    closing_inventory = opening_inventory + purchases - sales

    question = (
        f"Step 1. {company_name} starts the year with an inventory worth ${opening_inventory}. During the year, the company purchases "
        f"an additional ${purchases} worth of goods and sells ${sales} worth of inventory. Calculate the closing inventory."
    )

    solution = (
        f"Closing Inventory = Opening Inventory + Purchases - Sales\n"
        f"                  = {opening_inventory} + {purchases} - {sales} = ${closing_inventory}"
    )

    return question, solution

# Template 5 (basic)
def template_retained_earnings():
    """
    Scenario: Calculating Retained Earnings
    """
    company_name = random.choice(company_names)
    opening_retained_earnings = random.randint(1000, 10000)
    net_profit = random.randint(5000, 30000)
    dividends = random.randint(2000, 15000)
    closing_retained_earnings = opening_retained_earnings + net_profit - dividends

    question = (
        f"{company_name} starts the year with ${opening_retained_earnings} in retained earnings. During the year, the company "
        f"earns a net profit of ${net_profit} and pays out ${dividends} in dividends. Calculate the retained earnings at the end "
        f"of the year."
    )

    solution = (
        f"Step 1. Retained Earnings at Year-End = Opening Retained Earnings + Net Profit - Dividends\n"
        f"                                      = {opening_retained_earnings} + {net_profit} - {dividends} = ${closing_retained_earnings}"
    )

    return question, solution

# Template 6 (intermediate)
def template_inventory_valuation():
    """
    Scenario: Company XYZ is evaluating its inventory using different valuation methods.
    """
    # Data generation
    company_name = random.choice(company_names)
    inventory_start = random.randint(500, 1000)
    purchases = random.randint(2000, 5000)
    sales = random.randint(1500, 3000)
    fifo_cogs = random.randint(1000, 2000)
    lifo_cogs = fifo_cogs + random.randint(-200, 200)
    wac_cogs = fifo_cogs + random.randint(-100, 100)

    question = (
        f"{company_name} started the year with an inventory worth ${inventory_start}. Over the year, it made purchases worth "
        f"${purchases} and sold goods costing ${sales} using different valuation methods. Under FIFO, the cost of goods sold "
        f"was ${fifo_cogs}, under LIFO it was ${lifo_cogs}, and under WAC it was ${wac_cogs}. Calculate the closing inventory "
        f"for each method."
    )

    solution = (
        f"Using the formula: Closing Inventory = Starting Inventory + Purchases - COGS\n"
        f"Step 1. FIFO: {inventory_start} + {purchases} - {fifo_cogs} = ${inventory_start + purchases - fifo_cogs}\n"
        f"Step 2. LIFO: {inventory_start} + {purchases} - {lifo_cogs} = ${inventory_start + purchases - lifo_cogs}\n"
        f"Step 3. WAC: {inventory_start} + {purchases} - {wac_cogs} = ${inventory_start + purchases - wac_cogs}"
    )

    return question, solution

# Template 7 (intermediate)
def template_deferred_tax():
    """
    Scenario: Deferred tax assets and liabilities.
    """
    company_name = random.choice(company_names)
    profit_before_tax = random.randint(10000, 50000)
    temporary_difference = random.randint(2000, 8000)
    tax_rate = random.randint(20, 30) / 100
    tax_base = profit_before_tax - temporary_difference
    deferred_tax = temporary_difference * tax_rate

    question = (
        f"{company_name} reported a profit before tax of ${profit_before_tax}. There was a temporary difference of "
        f"${temporary_difference} between the tax base and the carrying amount of an asset. If the tax rate is "
        f"{int(tax_rate * 100)}%, calculate the deferred tax asset or liability and the taxable profit."
    )

    solution = (
        f"Step 1: Taxable profit = Profit Before Tax - Temporary Difference\n"
        f"                       = {profit_before_tax} - {temporary_difference} = ${tax_base}\n"
        f"Step 2: Deferred Tax = Temporary Difference * Tax Rate\n"
        f"                     = {temporary_difference} * {tax_rate} = ${deferred_tax}"
    )

    return question, solution

# Template 8 (intermediate)
def template_revaluation_assets():
    """
    Scenario: Revaluation of a fixed asset.
    """
    original_value = random.randint(50000, 100000)
    revaluation_surplus = random.randint(10000, 20000)
    new_value = original_value + revaluation_surplus

    question = (
        f"A piece of machinery was originally valued at ${original_value}. After a professional appraisal, "
        f"its value increased by ${revaluation_surplus}. What is the new book value of the asset?"
    )

    solution = (
        f"Step 1: New Book Value = Original Value + Revaluation Surplus\n"
        f"                       = {original_value} + {revaluation_surplus} = ${new_value}"
    )

    return question, solution

# Template 9 (intermediate)
def template_working_capital_ratio():
    """
    Scenario: Calculating working capital ratio.
    """
    company_name = random.choice(company_names)
    current_assets = random.randint(10000, 50000)
    current_liabilities = random.randint(5000, 25000)
    working_capital_ratio = current_assets / current_liabilities

    question = (
        f"{company_name} has current assets of ${current_assets} and current liabilities of ${current_liabilities}. "
        f"Calculate its working capital ratio and interpret whether it is favorable or not."
    )

    solution = (
        f"Step 1: Working Capital Ratio = Current Assets / Current Liabilities\n"
        f"                              = {current_assets} / {current_liabilities} = {working_capital_ratio:.2f}\n"
        f"Interpretation: A ratio above 1 indicates favorable liquidity, while below 1 indicates potential liquidity issues."
    )

    return question, solution

# Template 10 (intermediate)
def template_debt_equity_ratio():
    """
    Scenario: Calculating the debt-to-equity ratio.
    """
    company_name = random.choice(company_names)
    total_liabilities = random.randint(50000, 200000)
    shareholders_equity = random.randint(50000, 150000)
    debt_equity_ratio = total_liabilities / shareholders_equity

    question = (
        f"{company_name} has total liabilities of ${total_liabilities} and shareholders’ equity of ${shareholders_equity}. "
        f"Calculate its debt-to-equity ratio and comment on the company’s financial leverage."
    )

    solution = (
        f"Step 1: Debt-to-Equity Ratio = Total Liabilities / Shareholders’ Equity\n"
        f"                     = {total_liabilities} / {shareholders_equity} = {debt_equity_ratio:.2f}\n"
        f"Interpretation: A higher ratio indicates greater financial leverage and potentially higher financial risk."
    )

    return question, solution

# Template 11 (advanced)
def template_currency_translation():
    """
    Scenario: Cross-Border Acquisition and Currency Translation
    """
    company_name = random.choice(company_names)
    purchase_price_eur = random.randint(4000000, 6000000)
    net_assets_eur = random.randint(3000000, 4500000)
    exchange_rate = 1.2  # USD per EUR
    inflation_adjustment = random.randint(500000, 1000000)
    revaluation_surplus_eur = random.randint(300000, 600000)

    purchase_price_usd = purchase_price_eur * exchange_rate
    net_assets_usd = net_assets_eur * exchange_rate
    goodwill_usd = purchase_price_usd - net_assets_usd
    revaluation_surplus_usd = revaluation_surplus_eur * exchange_rate
    adjusted_assets_usd = net_assets_usd + revaluation_surplus_usd + inflation_adjustment

    question = (
        f"{company_name}, based in the USA, acquired a European subsidiary for €{purchase_price_eur}. At the acquisition date, the "
        f"exchange rate was 1 EUR = {exchange_rate} USD. The net assets of the subsidiary were valued at €{net_assets_eur}, and "
        f"a revaluation surplus of €{revaluation_surplus_eur} was identified. Additionally, inflation adjustments of ${inflation_adjustment} "
        f"were applied to the subsidiary's assets. Calculate the goodwill in USD, the revaluation surplus in USD, and the adjusted "
        f"total assets of the subsidiary after these adjustments."
    )

    solution = (
        f"Step 1. Purchase Price in USD = Purchase Price in EUR * Exchange Rate\n"
        f"                              = €{purchase_price_eur} * {exchange_rate} = ${purchase_price_usd}\n"
        f"Step 2. Net Assets in USD = Net Assets in EUR * Exchange Rate\n"
        f"                          = €{net_assets_eur} * {exchange_rate} = ${net_assets_usd}\n"
        f"Step 3. Goodwill in USD = Purchase Price in USD - Net Assets in USD\n"
        f"                        = ${purchase_price_usd} - ${net_assets_usd} = ${goodwill_usd}\n"
        f"Step 4. Revaluation Surplus in USD = Revaluation Surplus in EUR * Exchange Rate\n"
        f"                                   = €{revaluation_surplus_eur} * {exchange_rate} = ${revaluation_surplus_usd}\n"
        f"Step 5. Adjusted Total Assets = Net Assets in USD + Revaluation Surplus in USD + Inflation Adjustment\n"
        f"                              = ${net_assets_usd} + ${revaluation_surplus_usd} + ${inflation_adjustment} = ${adjusted_assets_usd}"
    )

    return question, solution

# Template 12 (advanced)
def template_hybrid_instruments():
    """
    Scenario: Structured Finance and Hybrid Instruments
    """
    bank_name = random.choice(bank_names)
    issuance_value = 10000000  # Total value of the hybrid instrument
    equity_component = 3000000  # Equity portion
    debt_component = issuance_value - equity_component
    interest_rate = 0.05  # 5% annual interest
    embedded_derivative_increase = random.randint(100000, 500000)
    equity_exercised = equity_component * 0.5
    remaining_equity = equity_component - equity_exercised
    accrued_interest = debt_component * interest_rate
    fair_value_impact = embedded_derivative_increase

    question = (
        f"{bank_name} issued a hybrid financial instrument worth ${issuance_value}, combining equity and debt components. "
        f"The equity component was valued at ${equity_component}, including a conversion option. The debt component, with a "
        f"fixed interest rate of {interest_rate * 100}%, was ${debt_component}. During the first year, 50% of the equity portion "
        f"was exercised, and the fair value of an embedded derivative increased by ${fair_value_impact}. Calculate (1) the remaining "
        f"equity component, (2) the debt liability after interest accrual, and (3) the fair value impact on the financial statements."
    )

    solution = (
        f"Step 1. Remaining Equity Component = Initial Equity Component - Exercised Equity\n"
        f"                                   = ${equity_component} - ${equity_exercised} = ${remaining_equity}\n"
        f"Step 2. Debt Liability After Interest Accrual = Debt Component + Accrued Interest\n"
        f"                                              = ${debt_component} + (${debt_component} * {interest_rate}) = ${debt_component + accrued_interest}\n"
        f"Step 3. Fair Value Impact = Increase in Fair Value of Embedded Derivative\n"
        f"                          = ${fair_value_impact}"
    )

    return question, solution

# Template 13 (advanced)
def template_pension_obligations():
    """
    Scenario: Pension Obligations and Actuarial Gains/Losses
    """
    company_name = random.choice(company_names)
    pension_obligation = 10000000  # Present value of pension obligation at the start of the year
    plan_assets = 8000000  # Fair value of plan assets at the start of the year
    contributions = 1000000  # Contributions made during the year
    benefits_paid = 500000  # Benefits paid during the year
    actuarial_loss = 600000  # Actuarial loss due to changes in discount rates
    expected_return = 400000  # Expected return on plan assets
    actual_return = 300000  # Actual return on plan assets

    net_pension_liability_end = (
        pension_obligation + actuarial_loss - plan_assets - actual_return + benefits_paid - contributions
    )

    question = (
        f"{company_name} sponsors a defined benefit pension plan. At the beginning of the year, the present value of the pension "
        f"obligation was ${pension_obligation}, and the fair value of plan assets was ${plan_assets}. During the year, contributions "
        f"of ${contributions} were made, benefits of ${benefits_paid} were paid, and an actuarial loss of ${actuarial_loss} was identified. "
        f"The expected return on plan assets was ${expected_return}, but the actual return was ${actual_return}. Calculate the "
        f"net pension liability at the end of the year."
    )

    solution = (
        f"Step 1. Net Pension Liability at Year-End = Pension Obligation + Actuarial Loss - Plan Assets - Actual Return\n"
        f"                                            + Benefits Paid - Contributions\n"
        f"                                          = {pension_obligation} + {actuarial_loss} - {plan_assets} - {actual_return} + {benefits_paid} - {contributions}\n"
        f"                                          = ${net_pension_liability_end}"
    )

    return question, solution

# Template 14 (advanced)
def template_derivatives():
    """
    Scenario: Accounting for Derivatives and Risk Management
    """
    company_name = random.choice(company_names)
    forward_contract_value = -50000  # Loss on the forward contract
    contract_notional = 1000000  # Notional value of the forward contract in EUR
    initial_rate = 1.1  # Initial exchange rate (USD per EUR)
    year_end_rate = 1.15  # Year-end exchange rate (USD per EUR)
    change_in_fair_value = forward_contract_value

    question = (
        f"{company_name} entered a forward contract to buy €{contract_notional} at an exchange rate of {initial_rate} USD per EUR. "
        f"At year-end, the exchange rate was {year_end_rate} USD per EUR, resulting in a fair value loss of ${-forward_contract_value}. "
        f"The company applies hedge accounting under IFRS 9. Calculate the impact of the hedge on the financial statements, "
        f"including the profit and loss (P&L) account and other comprehensive income (OCI)."
    )

    solution = (
        f"Step 1. Change in Fair Value of Forward Contract = {change_in_fair_value}\n"
        f"Step 2. Impact on P&L = {change_in_fair_value}\n"
        f"Step 3. Since hedge accounting is applied, the effective portion of the hedge would be reported in OCI, and the ineffective portion, "
        f"if any, would be reported in P&L. In this case, the entire loss of ${-forward_contract_value} impacts OCI as an effective hedge."
    )

    return question, solution

# Template 15 (advanced)
def template_business_combination():
    """
    Scenario: Business Combination and Non-Controlling Interest (NCI)
    """
    company_name1, company_name2 = random.sample(company_names, 2)
    purchase_price = 20000000  # Price paid for the acquisition
    net_assets_value = 15000000  # Fair value of the net assets
    nci_value = 5000000  # Fair value of the non-controlling interest (NCI)
    goodwill = purchase_price + nci_value - net_assets_value
    net_profit = 2000000  # Net profit of the subsidiary
    parent_share = 0.8  # Parent's ownership percentage
    parent_profit_share = net_profit * parent_share
    nci_profit_share = net_profit * (1 - parent_share)

    question = (
        f"{company_name1} acquires an 80% stake in {company_name2} for ${purchase_price}. The fair value of {company_name2}’s net identifiable assets "
        f"is ${net_assets_value}, and the fair value of the non-controlling interest (NCI) is ${nci_value}. During the year, {company_name2} "
        f"reported a net profit of ${net_profit}. Calculate the goodwill arising from the acquisition, the parent company’s share "
        f"of net profit, and the share of net profit attributable to NCI."
    )

    solution = (
        f"Step 1. Goodwill = Purchase Price + NCI Value - Net Assets Value\n"
        f"   = {purchase_price} + {nci_value} - {net_assets_value} = ${goodwill}\n"
        f"Step 2. Parent Company’s Share of Net Profit = Net Profit * Parent’s Ownership Percentage\n"
        f"   = {net_profit} * {parent_share} = ${parent_profit_share}\n"
        f"Step 3. NCI’s Share of Net Profit = Net Profit * (1 - Parent’s Ownership Percentage)\n"
        f"   = {net_profit} * {1 - parent_share} = ${nci_profit_share}"
    )

    return question, solution

# Example of generating a question
if __name__ == "__main__":
    for generate_fn in [
        template_cash_accounts_payable,
        template_balance_sheet_equation,
        template_depreciation_fixed_asset,
        template_retained_earnings,
        template_retained_earnings,
        template_inventory_valuation,
        template_deferred_tax,
        template_revaluation_assets,
        template_working_capital_ratio,
        template_debt_equity_ratio,
        template_currency_translation,
        template_hybrid_instruments,
        template_pension_obligations,
        template_derivatives,
        template_business_combination
    ]:
        question, solution = generate_fn()
        print("Question:", question)
        print("Solution:", solution)

        print("-" * 50)  # Separator for clarity
```
